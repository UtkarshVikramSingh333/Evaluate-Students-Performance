```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_data = pd.read_csv("StudentsPerformance (1)(4044).csv", sep=';', header=0)
df_data.head(5)
```

```python
#tabular information
df_data.info()
```

```python
#description of data 
df_data.describe()
```

```python
#correlation between numerical data
sns.heatmap(df_data.corr(), annot=True)
plt.show()
```

```python
#searching for any null values in the given data
df_data.isnull().sum()
```

```python
#selecting the variable 'race/ethnicity', and observing it's unique categories
df_data['race/ethnicity'].unique()
```

```python
#plotting number of students according to race/ethnicity
from matplotlib.pyplot import xticks
sns.countplot(x = 'race/ethnicity', data = df_data)
xticks(rotation = 90)
plt.xlabel('Race/Ethnicity')
plt.ylabel('Frequency')
plt.title('Race/Ethnicity')
```

```python
#according to the above graph, the maximum number of students fall under the group C category
```

```python
#observing the unique categories of the variable 'parental level of education'
df_data["parental level of education"].unique()
```

```python
#plotting according to the parental levels of education
from matplotlib.pyplot import xticks
sns.countplot(x="parental level of education", hue = "gender", data = df_data)
xticks(rotation=90)
plt.xlabel('Education Levels')
plt.ylabel('Frequency')
plt.title("Parental levels of Eductaion")
```

```python
#from the above graph, female students have better marks than male ones, considering the parental level of education
#the only category which has an exception is of 'high school', where male students perform better
```

```python
#observing unique lunch categories
df_data["lunch"].unique()
```

```python
#plotting according to the unique lunch categories
from matplotlib.pyplot import xticks
sns.countplot(x="lunch", hue = "gender", data = df_data)
xticks(rotation=90)
plt.xlabel('Different types of Lunch')
plt.ylabel('Frequency')
plt.title("Lunch")
```

```python
#plotting graph using parental level of education and lunch
from matplotlib.pyplot import xticks
sns.countplot(x="lunch", hue = 'parental level of education', data = df_data)
xticks(rotation=90)
plt.xlabel('Different types of Lunch')
plt.ylabel('Frequency')
plt.title("Lunch based on Parental Levels of Education")
```

```python
#plotting graph based on test preparation course and lunch
from matplotlib.pyplot import xticks
sns.countplot(x="test preparation course", hue = "lunch", data = df_data)
xticks(rotation=90)
plt.xlabel('Preparation Course')
plt.ylabel('Frequency')
```

```python
#plotting a graph based on test preparation course and parental level of education
from matplotlib.pyplot import xticks
sns.countplot(x="test preparation course", hue = "parental level of education", data = df_data)
xticks(rotation=90)
plt.xlabel('Preparation Course')
plt.ylabel('Frequency')
```

```python
#COMPARISON OF DIFFERENT SCORES
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.distplot(df_data['math score'], kde = False, color = 'c', bins = 30)
plt.ylabel("Frequency")
plt.title('Math Score Distribution')

plt.subplot(1,3,2)
sns.distplot(df_data['reading score'], kde = False, color ='c', bins = 30)
plt.ylabel("Frequency")
plt.title('Reading Score Distribution')

plt.subplot(1,3,3)
sns.distplot(df_data['writing score'], kde = False, color = 'c', bins = 30)
plt.ylabel("Frequency")
plt.title('Writing Score Distribution')
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
day_wise=df_data.groupby('gender')['math score'].mean().plot.bar()
plt.title("math score based on gender")
plt.ylabel('average math score', fontsize=12)
plt.xlabel('gender', fontsize=12)

plt.subplot(1,3,2)
day_wise=df_data.groupby('gender')['reading score'].mean().plot.bar()
plt.title("reading score based on gender")
plt.ylabel('average reading score', fontsize=12)
plt.xlabel('gender', fontsize=12)

plt.subplot(1,3,3)
day_wise=df_data.groupby('gender')['writing score'].mean().plot.bar()
plt.title("writing score based on gender")
plt.ylabel('average writing score', fontsize=12)
plt.xlabel('gender', fontsize=12)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.boxplot(x=df_data['gender'], y=df_data['math score'])

plt.subplot(1,3,2)
sns.boxplot(x=df_data['gender'], y=df_data['reading score'])

plt.subplot(1,3,3)
sns.boxplot(x=df_data['gender'], y=df_data['writing score'])
```
